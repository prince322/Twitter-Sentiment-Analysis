```python
y_train=np.array(train_tweets.airline_sentiment)
x_train=np.array(train_tweets.text)
x_test=np.array(test_tweets.text)
```

```python
from nltk.corpus import wordnet
def get_sample(tag):
    if tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('R'):
        return wordnet.ADV
    elif tag.startswith('V'):
        return wordnet.VERB
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
```

```python
from nltk.corpus import stopwords
import string
stop=stopwords.words('english')
stop+=list(string.punctuation)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
def clean_twitter(word):
    from nltk.tokenize import word_tokenize
    final_sentence=[]
    d=""
    for w in range(len(word)):
        c=word_tokenize(word[w])         #this will make a seprate string of each element in a document at given index
        c=c[2:]
        for j in c:
            if j.lower() not in stop:
                pos=pos_tag([j])
                clean=lemmatizer.lemmatize(j,pos=get_sample(pos[0][1]))
                d=d+" "+clean
        final_sentence.append(d.lower())        #After removing stopword it will add the reamining text of that document to the list
        d=""
    return final_sentence
```

```python
import re
c_data=clean_twitter(x_train)
```

```python
c_data
```

```python
count_vc=CountVectorizer(max_features=2300)
tf_vc=TfidfVectorizer(max_features=2000)
x_train_features=count_vc.fit_transform(c_data)
```

```python
x_test_features=count_vc.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
svc=SVC()
gr={'C':[1e2,1e3,5e3,1e4,5e4,1e5],'gamma':[1e-3,5e-4,1e-4,5e-3]}
abc=GridSearchCV(svc,gr)
```

```python
abc.fit(x_train_features,y_train)
y_pred=abc.predict(x_test_features)
y_pred
```
